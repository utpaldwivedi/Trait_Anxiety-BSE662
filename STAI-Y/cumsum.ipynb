```python
import pandas as pd

# Assuming your data is in a DataFrame named df
# First, sort the DataFrame by some index if not already sorted

# Drop the header row if it's not needed in calculations


# Reset the index


# Calculate the cumulative sum for the economic_score column
df['cumulative_economic_score'] = df['economic_score'].cumsum()

# Initialize a variable to keep track of the start index
start_index = 0

# Loop until you reach 750 entries
while start_index < 750:
    # Calculate the end index for the next 50 entries
    end_index = start_index + 50
    
    # Calculate the cumulative sum for the next 50 entries starting from 0
    df.loc[start_index:end_index, 'cumulative_economic_score_50'] = df.loc[start_index:end_index, 'economic_score'].cumsum()
    
    # Move to the next 50 entries
    start_index = end_index

# Fill NaN values in the new column with 0
df['cumulative_economic_score_50'].fillna(0, inplace=True)

# Round off the cumulative economic scores to 4 decimal places
df['cumulative_economic_score_50'] = df['cumulative_economic_score_50'].round(4)
```
